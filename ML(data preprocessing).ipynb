```python
#pandas.read-csv.
import pandas as pd
```

```python
pdDf=pd.read_csv('real_estate.csv')
pdDf.head()
```

```python
#Numpy.genfromtxt.
import numpy as np
data=np.genfromtxt('real_estate.csv',delimiter=',')
data
pd.DataFrame(data)
```

```python
#Q1) Handling missing Values.
df=pd.DataFrame(data)
df
```

```python
#part1 = isnull()
df.isnull()
```

```python
#part2 check for missing values in a pandas Series or DataFrame. It returns a boolean Series or DataFrame, where True indicates non-missing values and False indicates missing values.
df.notnull()
```

```python
#part4 = dropna()
# Understanding the Pandas .dropna() Method
import pandas as pd
df = pd.DataFrame(pdDf)

#df.dropna()
```

```python
#part3 = filling null value using fillna() fuction
df = pd.DataFrame(pdDf)
df.fillna(method='bfill')
```

```python
#part4 dropna() = Drops rows or columns containing missing values based on custom criteria.
df = pd.DataFrame(pdDf)
# Removing rows with missing values 
df_cleaned = df.dropna() 
# Displaying the DataFrame after removing missing values 
print("\nDataFrame after removing rows with missing values:") 
print(df_cleaned)
```

```python
#part5 = replace() = Replaces specific values with other values, facilitating data correction and standardization.
```

```python
#part6 = interpolate()
# Interpolation Techniques 
linear_interpolation = df['X1 transaction date'].interpolate(method='linear') 
quadratic_interpolation = df['X1 transaction date'].interpolate(method='quadratic') 

print("\nLinear Interpolation:") 
print(linear_interpolation) 

print("\nQuadratic Interpolation:") 
print(quadratic_interpolation)
```
